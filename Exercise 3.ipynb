```python
# Import the necessary libraries
import pandas as pd
```

```python
#Import the dataset and Assign it to a variable called users
users= pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|', index_col='user_id')
users
```

```python
#Discover what is the mean age per occupation 
users.age.mean() 
```

```python
# Discover the Male ratio per occupation and sort it from the most to the least
user_gender = users.groupby(['occupation','gender'])
user_gender = users.pivot_table(index='occupation', columns='gender', aggfunc=len, fill_value=0).age

user_gender['Total'] = user_gender['M']+user_gender['F']
user_gender['Male_Ratio'] = user_gender['M']/user_gender['Total']
user_gender = user_gender.sort_values(by=['Male_Ratio'],ascending=False)
user_gender
```

```python
# For each occupation, calculate the minimum and maximum ages
users.groupby('occupation').age.agg(['min', 'max'])
```

```python
# For each combination of occupation and gender, calculate the mean age
users.groupby(['occupation', 'gender']).age.mean()
```

```python
# For each occupation present the percentage of women and men
user_gender2 = users.groupby(['occupation','gender'])
user_gender2 = users.pivot_table(index='occupation', columns='gender', aggfunc=len, fill_value=0).age

user_gender2['Total'] = user_gender2['M']+user_gender2['F']
user_gender2['Male_Perc'] = user_gender2['M']/user_gender2['Total']
user_gender2['Female_Perc'] = user_gender2['F']/user_gender2['Total']
user_gender2
```
