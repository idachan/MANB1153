```python
# Import the necessary libraries
import pandas as pd
```

```python
#Import the dataset and Assign it to a variable called users
users= pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|', index_col='user_id')
users
```

```python
#Discover what is the mean age per occupation 
users.age.mean() 
```

```python
# For each occupation, calculate the minimum and maximum ages
users.groupby('occupation').age.agg(['min', 'max'])
```

```python
# For each combination of occupation and gender, calculate the mean age
users.groupby(['occupation', 'gender']).age.mean()
```

```python
# Male ratio per occupation with sorting it from the most to the least
users[users.gender == 'M'].groupby(['occupation', 'gender']).size()
```

```python
male=users[users.gender == 'M'].groupby(['occupation', 'gender']).size()
male
```

```python
gen_oc= users.groupby(['occupation']).size()
gen_oc
```

```python

(male/gen_oc*100).sort_values(ascending = False)
```
